```python
Q1-Write a Python program to check if a given string is an anagram.
def is_anagram(str1, str2):
    # Convert strings to lowercase and remove spaces
    str1 = str1.lower().replace(" ", "")
    str2 = str2.lower().replace(" ", "")

    # Check if the sorted characters of both strings are the same
    return sorted(str1) == sorted(str2)

# Example usage
string1 = "listen"
string2 = "silent"
print(is_anagram(string1, string2))  # Output: True

Q2-Write a Python program to find the maximum and minimum values in a dictionary.
def max_min_dict(dictionary):
    if len(dictionary) == 0:
        return None, None

    values = dictionary.values()
    return max(values), min(values)

# Example usage
my_dict = {'a': 5, 'b': 3, 'c': 9}
max_value, min_value = max_min_dict(my_dict)
print("Maximum value:", max_value)  # Output: 9
print("Minimum value:", min_value)  # Output: 3

Q3-Write a Python program to find the average of a list of numbers.
def average(numbers):
    if len(numbers) == 0:
        return None

    return sum(numbers) / len(numbers)

# Example usage
my_list = [1, 2, 3, 4, 5]
avg = average(my_list)
print("Average:", avg)  # Output: 3.0

Q4-Write a Python program to remove all vowels from a string.

def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return "".join(char for char in string if char not in vowels)

# Example usage
my_string = "Hello, World!"
new_string = remove_vowels(my_string)
print("New string:", new_string)  # Output: Hll, Wrld!

Q5-Write a Python function to calculate the factorial of a number using recursion.
python
Copy code
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

# Example usage
number = 5
factorial = factorial_recursive(number)
print("Factorial of", number, "is:", factorial)  # Output: 120

Q6-Write a Python program to merge two unsorted lists into a single sorted list using the merge sort algorithm.
def merge_sort_lists(list1, list2):
    merged_list = list1 + list2
    merged_list.sort()
    return merged_list

# Example usage
list_a = [5, 1, 4, 2]
list_b = [7, 3, 6]
sorted_list = merge_sort_lists(list_a, list_b)
print("Sorted list:", sorted_list)  # Output: [1, 2, 3, 4, 5, 6, 7]

Q7-Write a Python program to find the GCD (Greatest Common Divisor) of two numbers.
python
Copy code
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

# Example usage
num1 = 48
num2 = 36
result = gcd(num1, num2)
print("GCD of", num1, "and", num2, "is:", result)  # Output: 12
Q8-Write a Python program to find the second-largest element in a list.

def find_second_largest(numbers):
    if len(numbers) < 2:
        return None

    largest = second_largest = float('-inf')
    for num in numbers:
        if num > largest:
            second_largest = largest
            largest = num
        elif largest > num > second_largest:
            second_largest = num

    return second_largest

# Example usage
my_numbers = [9, 5, 2, 7, 1, 8]
second_largest = find_second_largest(my_numbers)
print("Second largest element:", second_largest)  # Output: 8

Q9-Write a Python program to find the longest common subsequence between two strings using dynamic programming.

def longest_common_subsequence(str1, str2):
    m = len(str1)
    n = len(str2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    lcs_length = dp[m][n]

    # Construct the longest common subsequence
    lcs = ""
    i = m
    j = n
    while i > 0 and j > 0:
        if str1[i - 1] == str2[j - 1]:
            lcs = str1[i - 1] + lcs
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1

    return lcs

# Example usage
string1 = "AGGTAB"
string2 = "GXTXAYB"
result = longest_common_subsequence(string1, string2)
print("Longest Common Subsequence:", result)  # Output: GTAB
```
